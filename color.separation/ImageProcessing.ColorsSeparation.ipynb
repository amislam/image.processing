<div style="direction:rtl">المهارة: فصل الألوان عن بعضها، لتحول الصورة الملونة لثلاثة صور، واحدة لكل لون</div>
<div style="direction:rtl">أهميتها: في عمليات الطباعة بالأحبار، تتم الطباعة ثلاثة مرات، مرة لكل لون، فيجب فصلهم</div>

```python
import matplotlib.pyplot as plt #نحتاج هذه المكتبة لطباعة الصور
import numpy as np #نحتاج هذه المكتبة للتعامل مع الأرقام
```

```python
x = plt.imread("./images.jpeg") #قراءة الصورة و وضع بياناتها في المتغير

plt.imshow(x); plt.show(); # اظهار الصورة
```

```python
print (x.shape) #طباعة ابعاد المتغير، و هو مصفوفة ثلاثية الأبعاد، 173 سطر، 292 عمود، 3مكونات لونية لكل نقطة،
```

```python
xr= np.copy(x)  # انشاء ثلاثة نسخ من المتغير، للون الاحمر و الأخضر و الأزرق علي  الترتيب
xg= np.copy(x)
xb= np.copy(x)
```

```python
xr[:,:,1:]=0 #لكل النقاط، ضع المكون الأخضر(1) والذي يليه و هو الأزرق(2) ، ضعهما بصفر، لتحصل على  المكون الأحمر فقط

plt.imshow(xr); plt.show(); # اظهار الصورة
```

```python
xg[:,:,0]=0 #لكل النقاط، ضع المكون الأحمر(0) والأزرق(2)بصفر، لتحصل على  المكون الأخضر فقط
xg[:,:,2]=0
plt.imshow(xg); plt.show(); # اظهار الصورة
```

```python
xb[:,:,0]=0 #لكل النقاط، ضع المكون الأحمر(0) الأخضر(1)بصفر، لتحصل على  المكون الأزرق فقط
xb[:,:,1]=0
plt.imshow(xb); plt.show(); # اظهار الصورة
```

```python
plt.imsave("./red.jpeg",    xr) # حفظ كل مكون في ملف منقصل
plt.imsave("./green.jpeg",  xg)
plt.imsave("./blue.jpeg",   xb)
```

```python
# اظهار جميع الصور: الأصلية، المكون الأحمر و الأخضر و الأزرق
plt.subplot(221)
plt.imshow(x)
plt.subplot(222)
plt.imshow(xr)
plt.subplot(223)
plt.imshow(xg)
plt.subplot(224)
plt.imshow(xb)
plt.show()
```
